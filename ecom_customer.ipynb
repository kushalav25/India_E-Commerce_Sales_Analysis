```python
import pandas as pd
df = pd.read_csv('india_ecommerce_customers.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['registration_date'] = pd.to_datetime(df['registration_date'],format = 'mixed')
df['registration_date'] = df['registration_date'].dt.strftime('%d-%m-%Y')
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
columns_strip = df[['name','gender','state','city','segment']]

for col in columns_strip:
    df[col]=df[col].astype(str).str.strip().str.title()
df.head()
```

```python
df['age'] = df['age'].astype(int)
df.head()
```

```python
gender_replace={
    'M':'Male',
    'F':'Female'
}

df['gender'] = df['gender'].replace(gender_replace)
df.head()
```

```python
from sqlalchemy import create_engine


username = "root"
password = "1234"
host = "localhost"
port = "3306"
database = "india_ecomm_sales"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

table_name = "customer" 
df.to_sql(table_name, engine, if_exists="replace", index=False)

pd.read_sql("SELECT * FROM customer LIMIT 5;", engine)
```
