```python
import pandas as pd
pd.set_option('display.max_column',None)
df = pd.read_csv("india_ecommerce_orders.csv")
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
date_fix = df[['order_date','delivery_date']]

for date in date_fix:
    df[date] = pd.to_datetime(df[date], format ='mixed')
    df[date] = df[date].dt.strftime('%d-%m-%Y')
df.head()
```

```python
text_col = df[['category','product','state','city','customer_segment']]

for col in text_col:
    df[col]=df[col].astype(str).str.strip().str.title()

df.head()
```

```python
df['platform'] = df['platform'].astype(str).str.strip().str.lower()
df.loc[df['platform'].str.contains('amazon'), 'platform'] = 'Amazon'
df.loc[df['platform'].str.contains('flipkart'), 'platform'] = 'Flipkart'
df['platform'] = df['platform'].str.title()
df.head()
```

```python
df['payment_method'] = df['payment_method'].astype(str).str.strip().str.lower()

pay_map ={
    'cash on delivery' : 'Cash On Delivery','debit card' : 'Debit Card','emi' : 'EMI','wallet' : 'Wallet','upi' :'UPI',
    'cc' : 'Credit Card','credit_card' : 'Credit Card' , 'credit card' : 'Credit Card',
    'net banking' : 'Net Banking','net_banking' : 'Net Banking','netbanking' : 'Net Banking'
}
df['payment_method'] = df['payment_method'].map(pay_map)
df.head()
```

```python
df['revenue'] = df['revenue'].astype(str).str.replace(',','', regex = False).str.strip()
df['revenue'] = pd.to_numeric(df['revenue'])
df.head()
```

```python
def dic_fix(value):

    if '%' in value:
        value = str(value).strip()
        return float(value.replace('%',''))/100
    return float(value)

df['discount_pct'] = df['discount_pct'].apply(dic_fix)

df.head()
```

```python
df['returned'] = df['returned'].astype(str).str.strip().str.lower()
ret_map = {'1' : 1, '0' :0,'yes' : 1, 'no' :0}
df['returned'] = df['returned'].map(ret_map).fillna(0).astype(int)

df
```
